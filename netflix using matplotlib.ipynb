```python
df["director"]=df["director"].fillna("Unknown")
df['country'] = df['country'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.columns
```

```python
content=df['type'].value_counts()
plt.bar(content.index, content.values)
plt.title("Movies vs TV Shows")
plt.show()
```

```python
yearly = df['release_year'].value_counts().sort_index()

plt.plot(yearly.index, yearly.values)
plt.title("Netflix Content Growth")
plt.show()
```

```python
country = df['country'].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.barh(country.index, country.values)
plt.title("Top 10 Countries")
plt.show()
```

```python
genres = df['listed_in'].str.split(', ').explode()

top_genres = genres.value_counts().head(10)

plt.barh(top_genres.index,
         top_genres.values)

plt.title("Top Genres")
plt.show()
```

```python
rating = df['rating'].value_counts()

plt.bar(rating.index,
        rating.values)

plt.xticks(rotation=45)
plt.title("Content Ratings")
plt.show()
```

```python
directors = df['director'].value_counts().head(10)

plt.barh(directors.index,
         directors.values)

plt.title("Top Directors")
plt.show()
```

```python
content = df['type'].value_counts()

plt.pie(content.values,
        labels=content.index)

plt.show()
```
